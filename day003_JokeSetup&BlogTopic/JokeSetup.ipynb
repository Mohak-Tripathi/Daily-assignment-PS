```python
from langchain_google_genai import ChatGoogleGenerativeAI
# Configure API key
import os
from getpass import getpass

if not os.environ.get("GOOGLE_API_KEY"):
    os.environ["GOOGLE_API_KEY"] = getpass("Enter your Google Gemini API key: ")


llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.1)
```

```python
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda

# 1A) Prompt to generate a clean, kid-friendly joke setup
setup_prompt = PromptTemplate.from_template(
    "Write a setup for a clean, kid-friendly joke. Keep it under 20 words."
)

# 1B) Chain: Prompt -> LLM -> String parser
setup_chain = setup_prompt | llm | StrOutputParser()

# Step B: explicit bridge (Runnable) — makes the contract crystal clear
# str  ->  {"setup": str}
to_setup_dict = RunnableLambda(lambda s: {"setup": s})

# (Optional quick test)
setup = setup_chain.invoke({})
print(setup)
```

```python
# Step 2: PromptTemplate + LCEL chain for the punchline

punchline_prompt = PromptTemplate.from_template(
    "Given this setup: {setup}\nWrite a witty, kid-friendly punchline in one short sentence."
)

punchline_chain = punchline_prompt | llm | StrOutputParser()

# final_chain = setup_chain | punchline_chain

# final_response  = final_chain.invoke({})

final_chain = setup_chain | to_setup_dict | punchline_chain

final_response = final_chain.invoke({})

print(f"""final response - {final_response}""")
```
