```python
from langchain_community.document_loaders import CSVLoader, PyMuPDFLoader, TextLoader
import os, tempfile

def load_documents(uploaded_files):
  docs = []


  for name in uploaded_files.keys():
    # print("name", name)
    ext = name.split(".")[-1]
    path = os.path.join(tempfile.gettempdir(), name)
    # print("path", path)

    with open(path, "wb") as file:
      file.write(uploaded_files[name]) # write the content of the files in this temp location

    if ext == "pdf":
      loader = PyMuPDFLoader(path)
    elif ext == "csv":
      loader = CSVLoader(path)
    elif ext == "txt":
      loader = TextLoader(path)
    else:
      print("unsupported file type")
      return

    file_docs = loader.load()
    # print("docs", docs)
    docs.extend(file_docs)

  return docs




# print(load_documents(uploaded))
```

```python
# This retriever is created through the instance of vector store and will return top 3 matching chunks after the sematic search
retriever = vector_store.as_retriever(search_kwargs={"k":3})
```

```python
# This function is just concatenating the top 3 relevant chunks returned from retriever and forming a single string that can
# be given to LLM as context
def format_chunks(relevant_chunks):
  return "\n\n".join(chunk.page_content for chunk in relevant_chunks)
```

```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_tavily import TavilySearch
tavily_key="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"

llm = GoogleGenerativeAI(model="gemini-2.5-flash", temperature=0, google_api_key="AIzaSyBjgjR2utN0tJneQ4P6mZ2Z4_1_N_02q30")

tool = TavilySearch(max_results=3, topic="general", tavily_api_key=tavily_key)
```

```python
from langchain_core.prompts import PromptTemplate

answer_prompt = PromptTemplate.from_template("""
You are a helpful assitant. Answer using ONLY the context below:

Content:
{context}

Question:
{question}

If the context is enough, answer accurately.
If not, repons only and exctly with this tag: [NEED_WEB_SEARCH], do not
provide any extra stuff just return the tag if the context is not enough
for answering the question.
""")

web_prompt = PromptTemplate.from_template("""
User Question: {question}

Web search results:
{web_results}

Based on the web results, provide a complete and helpful answer to the question asked by the user.
Provide citations as well, if available.
""")
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

# Answer the question based on context
determination_chain = (
    {
        "context": retriever | format_chunks,
        "question": RunnablePassthrough()
    }
    | answer_prompt
    | llm
    | StrOutputParser()
)

# if context is not enough then web search is required
web_search_chain = (
    {
        "question":RunnablePassthrough(),
        "web_results": lambda x: tool.invoke({"query": x})
    }
    |web_prompt
    | llm
    | StrOutputParser()
)
```

```python
def agent(question):
  response = determination_chain.invoke(question)

  if "[NEED_WEB_SEARCH]" in response:
    print("Info not found in the documents, web search is required....")
    return web_search_chain.invoke(question)
  else:
    return response
```

```python
while True:
    query = input("\n Ask a question (or type 'exit'): ")
    if query.lower() == "exit":
        break
    answer = agent(query)
    print("\n Answer:", answer)
```
