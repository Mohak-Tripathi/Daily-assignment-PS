```python
from getpass import getpass
import os

if not os.environ.get("GOOGLE_API_KEY"):
    os.environ["GOOGLE_API_KEY"] = getpass("Google api key")
```

```python
from langchain.chains import ConversationChain
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.3)

# response = llm.invoke("Hello")
# print(response)
```

```python
#memory section

from langchain.memory import ConversationBufferWindowMemory, ConversationSummaryBufferMemory, ConversationEntityMemory

# memory = ConversationBufferWindowMemory(k=2)

memory = ConversationSummaryBufferMemory(llm=llm, max_token_limit=500)


conversation = ConversationChain(
    llm=llm,
    verbose=True,
    memory=memory
)

while True:
  user_input  = input("User ")
  if user_input.lower() == "exit":
    break
  response = conversation.predict(input=user_input)
  print(response)


print("\n=== Summary Memory ===")
print(memory.moving_summary_buffer)  # or memory.load_memory_variables({})'



```

```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationEntityMemory

# Define a custom prompt template that includes 'entities'
template = """You are an AI assistant having a conversation with a user.
You have access to information about entities mentioned in the conversation:
{entities}

Previous conversation:
{history}
Human: {input}
AI:"""

prompt = PromptTemplate(
    input_variables=["entities", "history", "input"],
    template=template
)

# Initialize memory
memory = ConversationEntityMemory(llm=llm)

# Initialize the chain with the custom prompt and memory
conversation = LLMChain(
    llm=llm,
    prompt=prompt,
    verbose=True,
    memory=memory
)

while True:
  user_input  = input("User ")
  if user_input.lower() == "exit":
    break
  response = conversation.predict(input=user_input)
  print(response)

print("\n=== Entity Store ===")
# print(conversation.memory.store)  # shows extracted entities and facts
```
