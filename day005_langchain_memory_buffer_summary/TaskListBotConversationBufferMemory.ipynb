```python
import os
from getpass import getpass

if not os.environ.get("GOOGLE_API_KEY"):
    os.environ["GOOGLE_API_KEY"] = getpass("Enter your Google Gemini API key: ")
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash",temperature=0.1)

# smoke_test = llm.invoke("What is the capital of France?")
# print(smoke_test)
```

```python
#Memory and conversation chains

from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory


memory = ConversationBufferMemory()
conversation = ConversationChain(
    llm=llm,
    verbose=True,
    memory=memory
)


while True:
    user_input = input("User: ")
    if user_input.lower() == "exit":
        break
    response = conversation.predict(input=user_input)
    print("AI:", response)
```

```python
# Step : Print the conversation history stored in memory
print("\n=== Memory Buffer ===")
print(conversation.memory.buffer)
```
